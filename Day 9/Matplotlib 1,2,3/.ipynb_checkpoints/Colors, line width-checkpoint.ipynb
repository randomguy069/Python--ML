```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,5,11)
y = x**4
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color ='blue',linewidth=7,alpha=0.5) #adding color,linewidth or lw,alpha for transparency
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color ='blue',linewidth=7,linestyle='-.') #gives different shapes of line line
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color ='blue',linewidth=7,alpha=0.5,marker='*',markersize=10,markerfacecolor='black',markeredgewidth=3,markeredgecolor='gold') #marks the points on the plot
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color ='blue',linewidth=2,ls='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
