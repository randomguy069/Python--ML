```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure(figsize=(8,2)) #figure size (width , height)
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 2, ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
#saving a figure
fig.savefig('my_graph.png', dpi = 1000) #format can be anything from pdf to jpeg
```

```python
x = np.linspace(0,5,11)
y = x**3
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])

ax.plot(x,x ** 2,label='x squared') #adding a legend
ax.plot(x,x**3, label='x cubed')
ax.legend()#to avoid overlap
ax.legend(loc=0) #avoiding overlap of legend
ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')
```
