```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
x
```

```python
plt.plot(x,y) #basic plot
```

```python
plt.plot(x,y,'r-') #red color
```

```python
plt.plot(x,y,'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
#making multiple plots on the same canvas
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object oriented methods
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #adding axes.[left,bottom,height, width]
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title('Larger Plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2) #creates two sub plots. (One row with two subplots)
#axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 3, ncols=2) #3 rows and 2 columns of plots
```

```python
fig,axes =plt.subplots(nrows = 1, ncols =2)
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes =plt.subplots(nrows = 1, ncols =2)
axes[0].plot(x,y)
axes[0].set_title('First plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')
plt.tight_layout() #for tight layouts
```
